# Efficient Data Stream Anomaly Detection

This project demonstrates anomaly detection in real-time data streams, with a focus on temperature and energy price datasets. We use a Z-score method to detect anomalies, visualizations like histograms and boxplots for data distribution analysis, and a regression analysis to show the relationship between time and the detected anomalies.

## Data Overview

1. **Temperature Data** (`weather_features.csv`):
   - Source: A dataset of weather features, focusing on temperature.
   - Key Columns: 
     - `dt_iso`: Timestamps of temperature readings.
     - `temp`: Temperature values recorded at the given time.

2. **Energy Price Data** (`energy_dataset.csv`):
   - Source: A dataset of energy prices across time.
   - Key Columns:
     - `time`: Timestamps of energy price readings.
     - `price actual`: The actual price of energy at that time.

We will explore and analyze anomalies in both datasets, and visualize relationships using count plots and regression plots.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore
import logging

# Configure logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

class DataStreamAnomalyDetector:
    def __init__(self, data1, data2, threshold=3):
        self.data1 = data1
        self.data2 = data2
        self.threshold = threshold
        self.anomalies1 = []
        self.anomalies2 = []

    def detect_anomalies(self):
        logging.info("Starting anomaly detection for both datasets...")
        z_scores1 = zscore(self.data1)
        z_scores2 = zscore(self.data2)
        anomaly_indices1 = np.where(np.abs(z_scores1) > self.threshold)[0]
        anomaly_indices2 = np.where(np.abs(z_scores2) > self.threshold)[0]
        self.anomalies1 = anomaly_indices1
        self.anomalies2 = anomaly_indices2
        logging.info(f"Anomaly detection complete. {len(anomaly_indices1)} anomalies found in Dataset 1, "
                     f"{len(anomaly_indices2)} anomalies found in Dataset 2.")
        return anomaly_indices1, anomaly_indices2

    def visualize(self, timestamps1, timestamps2):
        plt.figure(figsize=(14, 8))
        plt.subplot(2, 1, 1)
        plt.plot(timestamps1, self.data1, label="Data Stream 1", color='blue')
        plt.scatter(timestamps1.iloc[self.anomalies1], self.data1.iloc[self.anomalies1], color='red', label='Anomalies', zorder=5)
        plt.title("Data Stream 1 (Temperature) with Anomalies")
        plt.xlabel("Timestamp")
        plt.ylabel("Value")
        plt.xticks(rotation=45)
        plt.legend()
        plt.grid(True)

        plt.subplot(2, 1, 2)
        plt.plot(timestamps2, self.data2, label="Data Stream 2", color='green')
        plt.scatter(timestamps2.iloc[self.anomalies2], self.data2.iloc[self.anomalies2], color='red', label='Anomalies', zorder=5)
        plt.title("Data Stream 2 (Price) with Anomalies")
        plt.xlabel("Timestamp")
        plt.ylabel("Value")
        plt.xticks(rotation=45)
        plt.legend()
        plt.grid(True)

        plt.tight_layout()
        plt.show()

    def visualize_histograms(self):
        plt.figure(figsize=(14, 6))
        plt.subplot(1, 2, 1)
        plt.hist(self.data1, bins=30, color='blue', alpha=0.7)
        plt.title("Histogram of Data Stream 1 (Temperature)")
        plt.xlabel("Value")
        plt.ylabel("Frequency")

        plt.subplot(1, 2, 2)
        plt.hist(self.data2, bins=30, color='green', alpha=0.7)
        plt.title("Histogram of Data Stream 2 (Price)")
        plt.xlabel("Value")
        plt.ylabel("Frequency")
        plt.tight_layout()
        plt.show()

    def visualize_boxplots(self):
        plt.figure(figsize=(14, 6))
        plt.subplot(1, 2, 1)
        plt.boxplot(self.data1, patch_artist=True, boxprops=dict(facecolor='blue', color='blue'))
        plt.title("Box Plot of Data Stream 1 (Temperature)")
        plt.ylabel("Value")

        plt.subplot(1, 2, 2)
        plt.boxplot(self.data2, patch_artist=True, boxprops=dict(facecolor='green', color='green'))
        plt.title("Box Plot of Data Stream 2 (Price)")
        plt.ylabel("Value")

        plt.tight_layout()
        plt.show()

    def visualize_heatmap(self, combined_data):
        plt.figure(figsize=(10, 8))
        sns.heatmap(combined_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
        plt.title("Correlation Heatmap of Features")
        plt.show()

    def log_anomalies(self, timestamps1, timestamps2):
        if len(self.anomalies1) > 0:
            logging.info("Detected anomalies in Dataset 1 at the following indices and timestamps:")
            for idx in self.anomalies1:
                logging.info(f"Index {idx}, Timestamp: {timestamps1.iloc[idx]}, Value: {self.data1.iloc[idx]}")
        else:
            logging.info("No anomalies detected in Dataset 1.")

        if len(self.anomalies2) > 0:
            logging.info("Detected anomalies in Dataset 2 at the following indices and timestamps:")
            for idx in self.anomalies2:
                logging.info(f"Index {idx}, Timestamp: {timestamps2.iloc[idx]}, Value: {self.data2.iloc[idx]}")
        else:
            logging.info("No anomalies detected in Dataset 2.")

    def visualize_countplot(self):
        """
        Visualize the count of anomalies in each dataset.
        """
        anomaly_counts = [len(self.anomalies1), len(self.anomalies2)]
        labels = ['Data Stream 1 (Temperature)', 'Data Stream 2 (Price)']

        plt.figure(figsize=(8, 6))
        sns.barplot(x=labels, y=anomaly_counts, palette="coolwarm")
        plt.title('Anomaly Counts in Data Streams')
        plt.ylabel('Number of Anomalies')
        plt.show()

    def visualize_regression(self, timestamps1, timestamps2):
        """
        Visualize regression between time and data streams and highlight anomalies.
        """
        plt.figure(figsize=(14, 8))

        # Regression plot for Data Stream 1 (Temperature)
        plt.subplot(2, 1, 1)
        sns.regplot(x=np.arange(len(self.data1)), y=self.data1, line_kws={"color": "red"})
        plt.scatter(self.anomalies1, self.data1.iloc[self.anomalies1], color='blue', zorder=5)
        plt.title("Regression Plot for Data Stream 1 (Temperature)")
        plt.xlabel("Index")
        plt.ylabel("Temperature")

        # Regression plot for Data Stream 2 (Price)
        plt.subplot(2, 1, 2)
        sns.regplot(x=np.arange(len(self.data2)), y=self.data2, line_kws={"color": "red"})
        plt.scatter(self.anomalies2, self.data2.iloc[self.anomalies2], color='blue', zorder=5)
        plt.title("Regression Plot for Data Stream 2 (Price)")
        plt.xlabel("Index")
        plt.ylabel("Price")

        plt.tight_layout()
        plt.show()
```
